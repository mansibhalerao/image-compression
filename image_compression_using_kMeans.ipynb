```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.image import imread

```

```python
#read the image using matplotlib
img=imread('D:/ML_datasets/image_compression/icecream.jpg')
plt.imshow(img)
plt.show()
print(img.shape)
print(img.size)
```

```python
img_shape=img.shape

#reshape it into 2D, where each row represents a pixel and the three coloum represent red,green,blue value.
X=img.reshape(img_shape[0]*img_shape[1],img_shape[2])
```

```python
#fit kmeans, with no. of clusters as the desired no. of colors.
kmeans=KMeans(n_clusters=7,n_init=10,max_iter=200)
kmeans.fit(X)
```

```python
##Assign colors to pixels based on their cluster center
#Each row in k_colors.cluster_centers_ represents the RGB value of a cluster centroid
#k_colors.labels_ contains the cluster that a pixel is assigned to
#The following assigns every pixel the color of the centroid it is assigned to

img_comp = kmeans.cluster_centers_[kmeans.labels_]

#reshape the image to origional dimensions.
img_comp = img_comp.reshape(img_shape[0],img_shape[1],img_shape[2])
```

```python
#display it
#the conversion to uint8 is needed due to some inconsistency in matplotlib.
plt.imshow(img_comp.astype(np.uint8))
plt.show()
```
